```python
# packages import
import pandas as pd
import numpy as np
import requests
from datetime import datetime as dt, timezone, timedelta
import json
import os

```

```python
# ===============================
# api request
# ===============================
data_url = "https://data.elexon.co.uk/bmrs/api/v1/balancing/pricing/market-index"

start_date = "2016-01-01"
from_date = ""

while from_date >= start_date:
    
   params = {
   "from": (dt.now(timezone.utc) - timedelta(days = 7)).strftime('%Y-%m-%dT00:00'),
   "to": dt.now(timezone.utc).strftime('%Y-%m-%dT00:00'),
   "dataProviders": "N2EXMIDP"
}

response = requests.get(data_url, params=params)
data = response.json()

with open("temp_result.json", "w")  as f:
   json.dump(data, f, indent=4)

os.system("code temp_result.json")

```
